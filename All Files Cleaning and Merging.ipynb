```python
import pandas as pd
import numpy as np
```

```python
# import 2019
citi_2019_df = pd.concat(map(pd.read_csv, [
                'raw_CSVs/JC-201901-citibike-tripdata.csv', 
                'raw_CSVs/JC-201902-citibike-tripdata.csv', 
                'raw_CSVs/JC-201903-citibike-tripdata.csv', 
                'raw_CSVs/JC-201904-citibike-tripdata.csv', 
                'raw_CSVs/JC-201905-citibike-tripdata.csv', 
                'raw_CSVs/JC-201906-citibike-tripdata.csv', 
                'raw_CSVs/JC-201907-citibike-tripdata.csv', 
                'raw_CSVs/JC-201908-citibike-tripdata.csv', 
                'raw_CSVs/JC-201909-citibike-tripdata.csv', 
                'raw_CSVs/JC-201910-citibike-tripdata.csv', 
                'raw_CSVs/JC-201911-citibike-tripdata.csv', 
                'raw_CSVs/JC-201912-citibike-tripdata.csv']))

# verify # of columns
citi_2019_df.head()
```

```python
# import all available 2020 CSVs. As of date of the analysis, data available through September 2020
citi_2020_df = pd.concat(map(pd.read_csv, [
                'raw_CSVs/JC-202001-citibike-tripdata.csv', 
                'raw_CSVs/JC-202002-citibike-tripdata.csv', 
                'raw_CSVs/JC-202003-citibike-tripdata.csv', 
                'raw_CSVs/JC-202004-citibike-tripdata.csv', 
                'raw_CSVs/JC-202005-citibike-tripdata.csv', 
                'raw_CSVs/JC-202006-citibike-tripdata.csv', 
                'raw_CSVs/JC-202007-citibike-tripdata.csv', 
                'raw_CSVs/JC-202008-citibike-tripdata.csv', 
                'raw_CSVs/JC-202009-citibike-tripdata.csv']))

# check data in dataframe
citi_2020_df.head()
```

```python
# merge all years into one dataframe
all_frames = [citi_2016_df, citi_2017_df, citi_2018_df, citi_2019_df, citi_2020_df]

citi_df = pd.concat(all_frames)
```

```python
# fill in the nan values in the birth year column
citi_df['birth year'] = citi_df['birth year'].fillna(0)
```

```python
# check for other nan values
null_df = citi_df[citi_df.isnull().any(axis=1)]
null_df.shape
```

```python
null_df.tail()
```

```python
# update nan user types to 'unknown'
citi_df['usertype'] = citi_df['usertype'].fillna('Unknown')
```

```python
# update "gender" to be more user-friendly, set as correct data type (object)
citi_df['gender'] = citi_df['gender'].map({1: 'Male', 2:'Female', 0:'Unknown'})
citi_df['gender'] = citi_df['gender'].astype('object')

# update birth year data type to 'int'
citi_df['birth year'] = citi_df['birth year'].astype('int')
```

```python
# clean outliers in tripduration
filtered_df = pd.DataFrame({'tripduration':np.random.normal(size=533)})
filtered_df = (citi_df[np.abs(citi_df.tripduration-citi_df.tripduration.mean()) <= (2*citi_df.tripduration.std())])

# check to see the data was removed
filtered_df.shape
```
